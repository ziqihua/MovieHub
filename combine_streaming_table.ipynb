```python
amazonprime_file = os.path.join(folder_path, "amazonprime_vs/titles.csv")
appletv_file = os.path.join(folder_path, "appletv_de/titles.csv")
disney_file = os.path.join(folder_path, "disney_vs/titles.csv")
hbo_file = os.path.join(folder_path, "hbo_vs/titles.csv")
hulu_file = os.path.join(folder_path, "hulu_vs/titles.csv")
netflix_file = os.path.join(folder_path, "netflix_vs/titles.csv")
paramount_file = os.path.join(folder_path, "paramount_vs/titles.csv")

# create a list of file paths
platform_df_list = []
filepath_list = [amazonprime_file, appletv_file, disney_file, hbo_file, hulu_file, netflix_file, paramount_file]
streaming_platform = ["Amazon Prime", "Apple TV+", "Disney+", "HBO Max", "Hulu", "Netflix", "Paramount+"]
try:
    assert len(filepath_list) == len(streaming_platform)
except AssertionError as error:
    print("AssertionError occured: Length of streaming platform name list is not equal to length of streaming platform data tables", error)


for i in range(len(filepath_list)):
    # print(filepath)
    filepath = filepath_list[i]
    cur_df = pd.read_csv(filepath)
    # print(cur_df.columns)

    # for each df, retain the following attributes
    # title, type, description
    # release_year, age_certification, runtime, genres
    # production_countries
    # imdb_id, imdb_score, imdb_votes

    # add the column name to the data table
    cur_df.insert(loc=len(cur_df.columns), column="platform", value = streaming_platform[i])

    # for each df, retain the following attributes
    # title, type, description
    # release_year, age_certification, runtime, genres
    # production_countries
    # imdb_id, imdb_score, imdb_votes
    cur_df_reduced = cur_df[['title', 'type', 'description', 'release_year', 'age_certification', 
                       'runtime', 'genres', 'production_countries', 'imdb_id', 'imdb_score', 
                       'imdb_votes', 'platform']]


    # add the final df to the list
    platform_df_list.append(cur_df_reduced)
    
```

```python
# combine all df
big_streaming_platform_df = pd.concat(platform_df_list, ignore_index=True)
```
